**RESPUESTA A PREGUNTA 6**

```python
# PARA CARGAR EL ARCHIVO DESDE UNA URL
# URL directa al archivo .csv calidad de vinos
import pandas as pd # Import the pandas library
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

# Leer el archivo CSV especificando el separador de punto y coma
VINOS = pd.read_csv(url, sep=';') #PANDAS PARA LEER LA BASE DE DATOS COMO HAY ; QUIERE DECIR QUE ESTA SEPARADOS POR ;

# Mostrar las primeras 5 filas para verificar que se cargó correctamente
print("Datos cargados exitosamente:")
print(VINOS.head())
```

```python
# EJERCICIO 6
# Reportar shape, columns, dtypes, head, describe
print("EJERCICIO 1: Exploración básica")
"filas y columnas"
print("shape:", VINOS.shape)
print("columnas:", list(VINOS.columns))
"nombres de columnas"
display(VINOS.dtypes)
"tipos de datos"
display(VINOS.head())
"primeras 5 filas"
display(VINOS.describe(include='all'))
"resumen estadístico"
```

**RESPUESTA A PREGUNTA 7**

```python
# EJERCICIO 7
# Dos variables numéricas: media, mediana, std y cuantiles (25%, 50%, 75%)
# Usaremos: 'alcohol' y 'pH'
# =========================
print("Estadísticos de dos variables numéricas")
vars_numericas = ['alcohol', 'pH']
estadisticos = []
for v in vars_numericas:
    serie = VINOS[v].dropna()
    estadisticos.append({
        'variable': v,
        'media': serie.mean(),
        'mediana': serie.median(),
        'desv_std': serie.std(),
        'Q25': serie.quantile(0.25),
        'Q50': serie.quantile(0.50),
        "igual a mediana"
        'Q75': serie.quantile(0.75),
        'min': serie.min(),
        'max': serie.max(),
        'n': serie.shape[0]
    })
df_estad = pd.DataFrame(estadisticos)
display(df_estad)
```

**RESPUESTA A PREGUNTA 8**

```python
# EJERCICIO 8
"Tabla de frecuencia categórica"
print("Tabla de frecuencias")
freq_abs = VINOS['quality'].value_counts().sort_index()
freq_rel = (VINOS['quality'].value_counts(normalize=True).sort_index() * 100).round(2)
tabla_freq = pd.DataFrame({'frecuencia': freq_abs, 'porcentaje_%': freq_rel})
display(tabla_freq)
```

**RESPUESTA A PREGUNTA 9**

```python
"FILAS FALTANTES"
"ejercicio 4"
print("FILAS FALTANTES")
faltantes_alcohol = pd.isnull(VINOS["alcohol"]).sum()
no_faltantes_alcohol = pd.notnull(VINOS["alcohol"]).sum()
print("alcohol → faltantes:", faltantes_alcohol, "| no faltantes:", no_faltantes_alcohol)
```
